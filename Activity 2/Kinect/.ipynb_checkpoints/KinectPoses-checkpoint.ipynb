```python
import pandas as pd
import numpy as np
import sys
import csv

student="S2"

"""
CSV File from parameter
"""
file_path = "data/"+student+".csv"

"""
Joint of Interest
"""
JOI = ['Head',
       'ShoulderCenter',
       'Spine',
       'HipCenter',
       'WristLeft',
       'HandLeft',
       'ElbowLeft',
       'ShoulderLeft',
       'ShoulderRight',
       'ElbowRight',
       'WristRight',
       'HandRight',
       ]



"""
"""
def convert_per_frame(df):
    joints={}   
    for i in range(df.__len__()):
        joint=df["Joint Type"][i]
        if not joint in joints.keys():
            values=[]
        else:
            values=joints[joint]
        x=df["Original Position X"][i]
        y=df[" Original Position Y"][i]
        values.append([x,y])
        joints[joint]=values
    return joints

"""
"""
def pose(i,data):
    verts[:] = np.array([data[j][i] for j in JOI])
    pose = get_pose_name(verts)
    #print 'frame: %d\npose: %s'%(i,pose)
    #label.set_text('frame: %d\npose: %s'%(i,pose))
    #if i == (kf_df.__len__() - 1):
            #plt.close()
            #sys.exit()
    return pose

"""
"""
def get_pose_name(skeleton):
    """
    Atributes from skeleton
    """
    d_HlSr = np.linalg.norm(verts[JOI.index('ShoulderRight')] - verts[JOI.index('HandLeft')])
    d_HlSl = np.linalg.norm(verts[JOI.index('ShoulderLeft')] - verts[JOI.index('HandLeft')])
    d_HrSr = np.linalg.norm(verts[JOI.index('ShoulderRight')] - verts[JOI.index('HandRight')])
    d_HrSl = np.linalg.norm(verts[JOI.index('ShoulderLeft')] - verts[JOI.index('HandRight')])
    d_HlHr = np.linalg.norm(verts[JOI.index('HandRight')] - verts[JOI.index('HandLeft')])
    d_ElEr = np.linalg.norm(verts[JOI.index('ElbowLeft')] - verts[JOI.index('ElbowRight')])
    d_HlB = np.linalg.norm(verts[JOI.index('Spine')] - verts[JOI.index('HandLeft')])
    d_HrB = np.linalg.norm(verts[JOI.index('Spine')] - verts[JOI.index('HandRight')])
    d_ElB = np.linalg.norm(verts[JOI.index('Spine')] - verts[JOI.index('ElbowLeft')])
    d_ErB = np.linalg.norm(verts[JOI.index('Spine')] - verts[JOI.index('ElbowRight')])
    v_ElHl = verts[JOI.index('ElbowLeft')] - verts[JOI.index('WristLeft')]
    v_ErHr = verts[JOI.index('ElbowRight')] - verts[JOI.index('WristRight')]
    v_SlEl = verts[JOI.index('ShoulderLeft')] - verts[JOI.index('ElbowLeft')]
    v_SrEr = verts[JOI.index('ShoulderRight')] - verts[JOI.index('ElbowRight')]
    o_ElHl = v_ElHl[1] / np.abs(v_ElHl[1]) if v_ElHl[1] != 0 else 0
    o_ErHr = v_ErHr[1] / np.abs(v_ErHr[1]) if v_ErHr[1] != 0 else 0
    o_SlEl = v_SlEl[1] / np.abs(v_SlEl[1]) if v_SlEl[1] != 0 else 0
    o_SrEr = v_SrEr[1] / np.abs(v_SrEr[1]) if v_SrEr[1] != 0 else 0

    """
    Rules determining poses
    """
    if (o_ElHl > 0 and o_ErHr > 0 and o_SlEl > 0 and o_SrEr > 0):#is_arms_down(skeleton):
        return 'arms down'
    elif (d_HlB > d_ElB and o_ElHl > 0) or (d_HrB > d_ErB and o_ErHr > 0):#is_pointing(skeleton):
        return 'pointing'
    elif (d_HlHr < d_ElEr) and (o_SlEl > 0) and (o_SrEr > 0) and (d_HlSr < d_HlSl):#is_crossed_arms(skeleton):
        return 'crossed arms'
    elif (o_ElHl < 0 and o_SlEl > 0) or (o_ErHr < 0 and o_SrEr > 0):#is_explaining(skeleton):
        return 'explaining'
    else:
        return 'no detected'
   
    
"""
Kinect Features Data Frame
"""
kf_df = pd.read_csv(file_path, sep=',', decimal='.')

joints=convert_per_frame(kf_df)
verts = np.zeros((JOI.__len__(),2))

with open("results/"+student+".csv", 'w') as csvfile:
    datawriter = csv.writer(csvfile, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    datawriter.writerow(["Time","Pose"])
    for i in range(joints['Head'].__len__()):
        datawriter.writerow([i/float(30),pose(i,joints)])
        
print("DONE")
```
